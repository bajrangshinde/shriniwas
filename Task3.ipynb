```python
import matplotlib.pyplot as plt
import pandas as pd

data = {
    'Month': ['January', 'February', 'March', 'April', 'May'],
    'Sales': [200, 300, 400, 350, 450],
    'Profit': [50, 80, 100, 90, 120]
}
df = pd.DataFrame(data)

plt.figure(figsize=(8, 5))  # Set figure size
plt.bar(df['Month'], df['Sales'], color='skyblue', label='Sales')

# Add title and labels
plt.title('Monthly Sales', fontsize=16)
plt.xlabel('Month', fontsize=12)
plt.ylabel('Sales ($)', fontsize=12)

# Add legend
plt.legend()

# Display the chart
plt.show()
```

```python
plt.figure(figsize=(8, 5))  # Set figure size
plt.plot(df['Month'], df['Profit'], marker='o', color='green', label='Profit')

# Add title and labels
plt.title('Monthly Profit', fontsize=16)
plt.xlabel('Month', fontsize=12)
plt.ylabel('Profit ($)', fontsize=12)

# Add legend
plt.legend()

# Display the chart
plt.show()
```

```python
plt.figure(figsize=(10, 6))

# Bar chart for Sales
plt.bar(df['Month'], df['Sales'], color='skyblue', label='Sales')

# Line chart for Profit
plt.plot(df['Month'], df['Profit'], marker='o', color='green', label='Profit')

# Add title and labels
plt.title('Monthly Sales and Profit', fontsize=16)
plt.xlabel('Month', fontsize=12)
plt.ylabel('Amount ($)', fontsize=12)

# Add legend
plt.legend()

# Display the chart
plt.show()
```
