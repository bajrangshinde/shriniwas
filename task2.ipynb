```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
data = pd.read_csv('INDIAvi.csv')

# Display the first few rows of the dataset
print("First 5 Rows of Data:")
display(data.head())

# Step 1: Handle Missing Values
# Check for any missing values
print("\nMissing Values per Column:")
missing_values = data.isnull().sum()
display(missing_values[missing_values > 0])

# Fill missing values with the mean (numerical) or mode (categorical)
for column in data.columns:
    if data[column].dtype == 'object':
        data[column] = data[column].fillna(data[column].mode()[0])
    else:
        data[column] = data[column].fillna(data[column].mean())

# Step 2: Summary Statistics
print("\nSummary Statistics:")
display(data.describe(include='all'))

print("\nDataset Info:")
data.info()

# Step 3: Univariate Analysis

# Numerical Variables: Detailed Histograms and KDE Plots
num_columns = data.select_dtypes(include=[np.number]).columns
for column in num_columns:
    plt.figure(figsize=(12, 6))
    sns.histplot(data[column], bins=50, kde=True, color='skyblue')
    sns.kdeplot(data[column], color="red", lw=2)  # Overlay KDE in red for detailed view
    plt.title(f'Histogram and KDE for {column} (Detailed)')
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.show()

    # Updated Distribution Plot using kdeplot with fill=True
    plt.figure(figsize=(12, 6))
    sns.kdeplot(data[column], fill=True, color="purple")  # Use fill=True instead of shade=True
    plt.title(f'Distribution Plot for {column}')
    plt.xlabel(column)
    plt.show()

# Enhanced Categorical Variables Analysis with Pie Charts and Bar Charts

cat_columns = data.select_dtypes(include=['object']).columns
for column in cat_columns:
    # Count Plot with Percentage Labels
    plt.figure(figsize=(12, 6))
    sns.countplot(x=column, data=data, palette="viridis", order=data[column].value_counts().index)
    plt.title(f'Count Plot for {column} with Percentages')
    plt.xlabel(column)
    plt.ylabel('Count')
    plt.xticks(rotation=45)
    for i, count in enumerate(data[column].value_counts().values):
        plt.text(i, count, f"{count}\n({count / len(data) * 100:.1f}%)", ha='center')
    plt.show()

    # Pie Chart for Categorical Data Distribution
    plt.figure(figsize=(8, 8))
    data[column].value_counts().plot.pie(autopct='%1.1f%%', colors=sns.color_palette("viridis", len(data[column].unique())))
    plt.title(f'Pie Chart of {column}')
    plt.ylabel('')  # Hide y-axis label for a cleaner look
    plt.show()

# Step 4: Bivariate Analysis

# Scatter Plot with Regression Line for Numerical Variables
for col_x in num_columns:
    for col_y in num_columns:
        if col_x != col_y:
            plt.figure(figsize=(10, 6))
            sns.regplot(x=data[col_x], y=data[col_y], scatter_kws={"s": 10}, line_kws={"color": "red"})
            plt.title(f'Regression Plot between {col_x} and {col_y}')
            plt.xlabel(col_x)
            plt.ylabel(col_y)
            plt.show()

# Boxen Plot for Categorical vs Numerical Variables
for column in cat_columns:
    for num_col in num_columns:
        plt.figure(figsize=(12, 6))
        sns.boxenplot(x=data[column], y=data[num_col], palette="pastel")
        plt.title(f'Boxen Plot of {num_col} by {column}')
        plt.xlabel(column)
        plt.ylabel(num_col)
        plt.xticks(rotation=45)
        plt.show()

# Swarm Plot for Detailed Categorical vs Numerical
for column in cat_columns:
    for num_col in num_columns:
        plt.figure(figsize=(12, 6))
        sns.swarmplot(x=data[column], y=data[num_col], palette="deep", size=3)
        plt.title(f'Swarm Plot of {num_col} by {column}')
        plt.xlabel(column)
        plt.ylabel(num_col)
        plt.xticks(rotation=45)
        plt.show()

# Step 5: Correlation Analysis

# Detailed Correlation Heatmap
plt.figure(figsize=(14, 10))
correlation_matrix = data[num_columns].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', square=True, annot_kws={"size": 8})
plt.title("Correlation Matrix of Numerical Variables (Detailed)")
plt.show()

# Step 6: Advanced Visualizations

# Violin Plot for Categorical vs Numerical with Swarm Overlay
for column in cat_columns:
    for num_col in num_columns:
        plt.figure(figsize=(12, 6))
        sns.violinplot(x=data[column], y=data[num_col], inner="quartile", palette="pastel")
        sns.swarmplot(x=data[column], y=data[num_col], color='black', size=2, alpha=0.5)
        plt.title(f'Violin Plot of {num_col} by {column}')
        plt.xlabel(column)
        plt.ylabel(num_col)
        plt.xticks(rotation=45)
        plt.show()

# Line Plot for Time-Series Data (if any ordered or date columns exist)
date_columns = [col for col in data.columns if 'date' in col.lower()]
for date_col in date_columns:
    plt.figure(figsize=(12, 6))
    sns.lineplot(data=data, x=date_col, y=num_columns[0])  # Plotting the first numerical column against Date
    plt.title(f'Line Plot of {num_columns[0]} over {date_col}')
    plt.xlabel(date_col)
    plt.ylabel(num_columns[0])
    plt.xticks(rotation=45)
    plt.show()
```
